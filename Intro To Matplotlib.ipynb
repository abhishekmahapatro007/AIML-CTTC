<h1 style = 'font-size: 20px; color : yellowgreen'><b><i><mark>MatPlotLib and Seaborn</mark></i></b><h1>
<pre style = 'font-size :14px'>Data Visualisation Techniques

```python
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
import pandas as pd
```

```python
x = np.random.randint(1,11,9)
y = np.random.randint(1,22,9)
plt.plot(x,y,marker = 'p',color = 'green',linestyle = 'dashed',markersize = 10)
plt.title("Relation Between Random Datasets")
plt.grid()
plt.xlabel('Random x axis')
plt.ylabel('Random y axis')
plt.show() #it's not necessary but consumes more memory
```

<h3><mark><b><i>Bar Graph </i></b></mark></h3>

```python
x = np.random.randint(1,10,7)
y = np.random.randint(1,25,7)
plt.bar(x,y,label = 'Blue randoms',color = 'blue')
plt.bar(y,x,label = 'Green randoms',color = 'green')
plt.title('Bar Chart of Random Datasets')
plt.xlabel('Random x-sets')
plt.ylabel('Random y-sets')
plt.legend()
```

<h3><mark><b><i>Histogram </i></b></mark></h3>
<pre>Shows The Frequency of a particular value

```python
interval = range(0,100,10)
age = np.random.randint(1,100,len(interval))
plt.hist(age,interval,histtype = 'bar',rwidth = 0.5)
#first parameter takes the dataset,2nd parameter shows the interval and it gives the frequency in o/p
#interval is not mandatory,u can skip it'll auto manage
#histtype is by default bar and we can ignore it as well
plt.title('Histogram')
plt.xlabel("Age of the population")
plt.ylabel('Age Interval')
plt.show()
```

<h3><mark><b><i>Scatter Plot </i></b></mark></h3>


```python
x = np.random.randint(1,800,40)
y = np.random.randint(1,1000,40)
plt.scatter(x,y,marker = '^',color = 'green')
plt.title('Scatter Plot')
plt.xlabel('X- Axis')
plt.ylabel('Y- Axis')
plt.show()
```

<h3><mark><b><i> Pie Chart</i></b></mark></h3>

```python
slices = [12,3,5,8,10]
activities = ['Phone','Study','Eat','Sleep','Gaming']
cols = ['r','y','m','b','g']
plt.pie(slices,labels = activities,colors = cols,startangle = 90,shadow = True,explode = (0,0,0.1,0,0),
autopct = '%1.1f%%')
#explode is amount the piece will come out,1.1f is to calculate percentage 
#and .1 is for decimal numbers after int,the integer doesn't matter we can use 0.1/1.1/2.1
plt.title('Pie Chart')
plt.show()
```

```python
#scatter-plot
iris.columns = ['SL' ,'SW','PL','PW','Type']
sb.scatterplot(x = iris.SL, y = iris.SW, marker = '*')
plt.show()
```

<b><i><mark>Distributed Plot</mark> is Combination of</i></b>
<ol>
    <li><mark>Displot</mark>
    <li><mark>KDEplot</mark> (Kernel Density Estimate)
</ol>

```python
sb.distplot(iris.SL)   #warning bcuz it'll get removed in future,but now the subsets
plt.show()
```

```python
sb.displot(iris.SL)  
plt.show()
```

```python
sb.kdeplot(iris.SL)  
plt.show()
```

```python
sb.distplot(iris.SL[iris['Type'] == 'Iris-setosa'],label = 'Iris Setosa',color = 'red')
sb.distplot(iris.SL[iris['Type'] == 'Iris-virginica'],label = 'Iris Virginica',color = 'yellowgreen')
sb.distplot(iris.SL[iris['Type'] == 'Iris-versicolor'],label = 'Iris Versicolor',color = 'blue')
plt.legend()
plt.show()
```

```python
#displot will plot in several graphs
sb.displot(iris.SL[iris['Type'] == 'Iris-setosa'],color = 'red')
sb.displot(iris.SL[iris['Type'] == 'Iris-virginica'],color = 'yellowgreen')
sb.displot(iris.SL[iris['Type'] == 'Iris-versicolor'],color = 'blue')
plt.show()
```

```python
#We can plot kde in 1 graph
sb.kdeplot(iris.SL[iris['Type'] == 'Iris-setosa'],color = 'red')
sb.kdeplot(iris.SL[iris['Type'] == 'Iris-virginica'],color = 'yellowgreen')
sb.kdeplot(iris.SL[iris['Type'] == 'Iris-versicolor'],color = 'blue')
plt.show()
```

<h3><mark><b><i> Countplot </i></b></mark></h3>
<pre> Returns the count of particular Category

```python
sb.countplot(x = iris['Type'])
plt.show()
```

<h3><mark><b><i> Box Plot</i></b></mark></h3>
<pre>Categorial data will be in x axis

```python
sb.boxplot(x = iris['Type'],y = iris.SL)
plt.show()
```

<mark>Outliers</mark>
<pre>Will show the extreme data,if the dataset don't differ that much outlier will be low or 0

<h3><mark><b><i>Swarm Plot </i></b></mark></h3>
<pre>A swarm plot is a scatter plot with non-overlapping points that represent categorical data

```python
sb.swarmplot(x = iris['Type'],y = iris.SL)  #u can't use iris.Type bcuz of the type method
plt.show()
```

<h3><mark><b><i>Violin plot </i></b></mark></h3>

```python
sb.violinplot(x = iris['Type'],y = iris.SL)
plt.show()
```

<h3><mark><b><i>Pair plot </i></b></mark></h3>
<pre> Will return every possible graphs

```python
sb.pairplot(iris)
plt.show()
```

<b>Use Of Hue</b>
<pre>Will show only differentiation of particular column

```python
sb.pairplot(iris,hue = 'Type') 
plt.show()
```

<h3><mark><b><i>Heat Map</i></b></mark></h3>
<pre>A representation of data in the form of a map or diagram in which data values are represented as colours


```python
c = iris.corr() #correlation
c
```

```python
sb.heatmap(c,annot = True,cmap = 'Blues') 
#cmap decide the color palette
#annot will show the values in the box
plt.show()
#if values are towards 1/-1 it's highly correlated
#if values are towards 0 it's less correlated
```
