# Auto-MPG

```python
auto = pd.read_csv(r"C:\Users\Lab-26\Downloads\auto-mpg.data",header=None,delim_whitespace=True)
auto
```

```python
auto.columns = ['mpg','cylinders','displacement',
               'horsepower','weight','acceleration',
               'model year','origin','car name']
auto.head()
```

```python
auto.isnull().sum()
```

```python
auto.dtypes
```

```python
for i in auto.columns:
    print(i,':','\n',auto[i].unique())
```

```python
for i in auto.columns:
    print(i,':',sum(auto[i]=='?'))
```

```python
d = auto.describe(include='all')
d
```

```python
#replacing the "?" mark
auto.horsepower.replace('?',d['horsepower'][2],inplace=True)
```

```python
for i in auto.columns:
    print(i,':',sum(auto[i]=='?'))
```

```python
auto.dtypes
```

```python
#to change data type we use astype
auto['horsepower'] = auto['horsepower'].astype('float64')
```

```python
auto.dtypes
```

```python
#to drop a column
auto.drop('car name',axis=1,inplace=True)
```

```python
auto.head()
```
