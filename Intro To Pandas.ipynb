Series From Dictionary

```python
dict = {'a':1,'b':2,'c':3}
print(dict)
type(dict)
```

```python
#key will be the index position since they'll are unique and the values will be the series
s = pd.Series(dict)
s
```

```python
type(s)
```

```python
index = ['d','e','f']
s1 = pd.Series(dict,index)
s1
#can't change the value because all the values are already assigned to particular keys and can't be overriden
```

```python
s2 = pd.Series(l,index = [10,10,20,30])
s2
```

```python
s2[10]   #we can store multiple values in a single index
```

```python
s[0]  #we can't access the values using index if there are duplicate keys
```
